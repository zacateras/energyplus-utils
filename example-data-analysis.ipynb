# Best practice for Python programming for data analysis

__Always use vectorized implementation for coding__

```python
def f(x, y):
    # A somewhat trivial example
    return x + y
```

```python
t0= time.time()

x =[]
number_of_data = 8760*10
for i in range(0,number_of_data):
    x.append(f(i,i*2))

t1 = time.time()-t0
print("Execution time: ",t1)
```

```python
t0= time.time()

x = np.linspace(0,number_of_data,num=number_of_data)
y = f(x,x*2)

t2 = time.time()-t0
print("Execution time: ",t2)
```

```python
speed_up = t1/t2
speed_up
```
